```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

for i in range(100):
    array = np.random.randint(0, 11, size=10000)
    cumsum = np.cumsum(array)
    results = np.empty(10000)
    for j in range(10000):
        results[j] = cumsum[j] / (j+1)
    plt.plot(np.arange(1, 10001), results)
    
plt.ylim(4, 7)
plt.grid(True, linestyle='--')
```
