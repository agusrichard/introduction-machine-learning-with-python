# Preprocessing and Scaling

__Import libraries__

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mglearn
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, Normalizer
```

Neural networks and SVMs are very sensitive to the scaling of the data. So we have to adjust the data so the model will perform better.

```python
mglearn.plots.plot_scaling()
```

## Different Kinds of Preprocessing

- The StandardScaler in scikit-learn ensures that for each feature the mean is 0 and the variance is 1, bringing all features to the same magnitude. Does not ensure any particular minimum and maximum values for the features.
- The RobustScaler works the same way as StandardScaler but uses median and quartiles. It ignores the datapoints that are very different from the rest (outliers).
- The MinMaxScaler shifts the data such that all features are exactly between 0 and 1. In two dimension this means all the data are contained inside the rectangle between 0 and 1 for both axis.
- The Normalizer scales each data point such that the feature vector has a Euclidean length of 1. In other words, it projects all data points on a circle or sphere or higher dimensional object). This normalization is often used when only the direction (or angle) of the data matters, not the length of the feature vector.

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=100, n_features=2)
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
std = StandardScaler()
X_std =  std.fit_transform(X)
plt.scatter(X_std[:, 0], X_std[:, 1], c=y)
plt.grid(True)
```

```python
norm = Normalizer()
X_norm = norm.fit_transform(X)
plt.scatter(X_norm[:, 0], X_norm[:, 1], c=y)
plt.grid(True)
```

## Applying Data Transformations

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, 
                                                    random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
# transform the data
X_train_scaled = scaler.transform(X_train)
# print dataset properties before and after scaling
print("transformed shape: {}".format(X_train_scaled.shape))
print("per-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))
print("per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling:\n {}".format(X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(X_train_scaled.max(axis=0)))
```

```python
# transform test set
X_test_scaled = scaler.transform(X_test)
# print test data properties after scaling
print("per-feature minimum after scaling:\n{}".format(X_test_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n{}".format(X_test_scaled.max(axis=0)))
```

Why some values on test set beyond the boundary? The explanation is that the MinMaxScaler (and all the other scalers) always applies exactly the same transformation to the training and the test set. This means the transform method always subtracts the training set minimum and divides by the training set range, which might be different from the minimum and range for the test set.

## Falacy: Scaling Training and Test Data the Same Way

```python
from sklearn.datasets import make_blobs
# make synthetic data
X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
# split it into training and test sets
X_train, X_test = train_test_split(X, random_state=5, test_size=.1)

# plot the training and test sets
fig, axes = plt.subplots(1, 3, figsize=(13, 4))

axes[0].scatter(X_train[:, 0], X_train[:, 1],
c=mglearn.cm2(0), label="Training set", s=60)
axes[0].scatter(X_test[:, 0], X_test[:, 1], marker='^',
c=mglearn.cm2(1), label="Test set", s=60)
axes[0].legend(loc='upper left')
axes[0].set_title("Original Data")

# scale the data using MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# visualize the properly scaled data
axes[1].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
c=mglearn.cm2(0), label="Training set", s=60)
axes[1].scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], marker='^',
c=mglearn.cm2(1), label="Test set", s=60)
axes[1].set_title("Scaled Data")

# rescale the test set separately
# so test set min is 0 and test set max is 1
# DO NOT DO THIS! For illustration purposes only.
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)
# visualize wrongly scaled data

axes[2].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1],
c=mglearn.cm2(0), label="training set", s=60)
axes[2].scatter(X_test_scaled_badly[:, 0], X_test_scaled_badly[:, 1],
marker='^', c=mglearn.cm2(1), label="test set", s=60)
axes[2].set_title("Improperly Scaled Data")
for ax in axes:
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
fig.tight_layout()
```

## The Effect of Preprocessing on Supervised Learning

```python
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, 
                                                    random_state=1)

svm = SVC(C=100)
svm.fit(X_train, y_train)
print("Test score accuracy: {}".format(svm.score(X_test, y_test)))
```

```python
# scaling
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# learning an SVM on the scaled training set
svm.fit(X_train_scaled, y_train)

# scoreing on the scaled test set
print("Scaled test_set accuracy: {}".format(svm.score(X_test_scaled, y_test)))
```

```python
# preprocessing using zero mean and unit variance scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# learning an SVM on the scaled training data
svm.fit(X_train_scaled, y_train)

# scoring on the scaled test set
print("SVM test accuracy: {}".format(svm.score(X_test_scaled, y_test)))
```

# Important Points

- Scaler:
    - StandardScaler: scale and centerize the data, use mean and standard deviation.
    - RobustScaler: same as StandardScaler, use median and quartiles.
    - MinMaxScaler: put the data on a rectangle (two dimensional data).
    - Normalize: projects it on cirles.
- Applying:
    - Fit and transform.
    - Ex: 
        - scaler.fit(X_train)
        - scaler.fit_transform(X_train) (returning arraylike)
        - scaler.transform(X_train) (returning arraylike)
- Falacy of thinking:
    - Fit the data on training set and transform it.
    - Transform the test set but don't fit with it.
