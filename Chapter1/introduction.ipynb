## Measuring Success: Training and Testing Data

Split our data data into training set and test set. We use training set to build our model and test set for validate our model. Deciding the size of our test set is a matter of the actual size of our dataset, but the good rule of thumb is 25% of total data.

Conventions, Data (feature matrix) denoted by X and label (target vector) denoted by y.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], 
    random_state=0, test_size=0.25)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

```python
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

## First Things First: Look at Your Data

```python
# create a DataFrame from data in X_train
iris_dataframe = pd.DataFrame(X_train,
                             columns=iris_dataset['feature_names'])
```

```python
iris_dataframe.shape
```

```python
# create a scatter matrix from the dataframe, color by y_train
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15),
                          marker='o', hist_kwds={'bins':20}, s=60,
                          alpha=0.8, cmap=mglearn.cm3);
```

```python
sns.pairplot(data=iris_dataframe);
```

## Building Your First Model: k-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train);
```

## Making Predictions

Imagine we found an iris in the wild with a sepal length of
5 cm, a sepal width of 2.9 cm, a petal length of 1 cm, and a petal width of 0.2 cm.
What species of iris would this be?

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
```

```python
X_new.shape
```

```python
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted target name: {}".format(
        iris_dataset['target_names'][prediction]))
```

## Evaluating the Model

```python
y_pred = knn.predict(X_test)
print("Test set predictions: \n{}".format(y_pred))
```

```python
print("Test set score: {:.5f}".format(knn.score(X_test, y_test)))
```

For this model, the test set accuracy is about 0.97, which means we made the right
prediction for 97% of the irises in the test set. Under some mathematical assumptions,
this means that we can expect our model to be correct 97% of the time for new
irises. For our hobby botanist application, this high level of accuracy means that our
model may be trustworthy enough to use. In later chapters we will discuss how we
can improve performance, and what caveats there are in tuning a model.
