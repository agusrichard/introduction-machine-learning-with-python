# Introduction to Supervised Learning

Supervised learning is used whenever we want to predict a certain outcome from a given input and we have examples of input/output paires. Our goal is to make accurate predictions for new, never-seen-before data.

## Classification and Regression

These are two major types of supervised machine learning problem. 

For Classification, our goal is to predict class label. Could be binary class (only two class/yes-no) or multiclass (more than two class)

For Regression, our goal is to predict a continuous number or a floating-point number. Ex: predicting a person's annual income from their education, their age, and where they live.

## Generalization, Overfitting and Underfitting

In supervised learning we build our model based on training set and try to make prediction to unseen new data. If a model is able to make accurate prediction on unseen data, we say it is able to __generalize__ from the training set to the test set.

Building a model that is too complex for the amount of
information we have,is called __overfitting__. Overfitting
occurs when you fit a model too closely to the particularities of the training set and obtain a model that works well on the training set but is not able to generalize to new data. 

On the other hand, if your model is too simple, then you might not be able to capture all the aspects of and variability
in the data, and your model will do badly even on the training set. Choosing too simple a model is called underfitting.
